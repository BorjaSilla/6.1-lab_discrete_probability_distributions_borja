## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
from scipy.stats import poisson
````

````python
# Given data
total_fruits = 100
apples = 60
oranges = 40

# Probability of picking an apple
probability_apple = apples / total_fruits

# Probability of picking an orange
probability_orange = oranges / total_fruits

print(f"Probability of picking an apple: {probability_apple:.2f} or {probability_apple * 100:.2f}%")
print(f"Probability of picking an orange: {probability_orange:.2f} or {probability_orange * 100:.2f}%")
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
probability_apple = 0.6
probability_orange = 0.4

# Probability that the first 5 fruits are all apples
prob_5_apples = probability_apple ** 5

# Probability that the first 5 fruits are all apples and the next 15 fruits are all oranges
prob_5_apples_15_oranges = prob_5_apples * (probability_orange ** 15)

print(f"Probability of the first 5 fruits being all apples: {prob_5_apples * 100:.4f}%")
print(f"Probability of the first 5 fruits being all apples and the next 15 being all oranges: {prob_5_apples_15_oranges * 100:.10f}%")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
probability_apple = 0.6
probability_orange = 0.4

# Number of trials (fruits picked)
n = 20

# Number of successes (apples)
k_apples = 5

# Binomial coefficient
binomial_coefficient = math.comb(n, k_apples)

# Probability of getting exactly 5 apples and 15 oranges
prob_5_apples_15_oranges_binomial = binomial_coefficient * (probability_apple ** k_apples) * (probability_orange ** (n - k_apples))

print(f"Probability of getting exactly 5 apples and 15 oranges: {prob_5_apples_15_oranges_binomial * 100:.10f}%")
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
probability_apple = 0.6
probability_orange = 0.4

# Number of trials (fruits picked)
n = 20

# Calculate cumulative probability of less than 5 apples using the binomial distribution
cumulative_prob_less_than_5_apples = binom.cdf(4, n, probability_apple)

print(f"Probability of less than 5 fruits picked being apples: {cumulative_prob_less_than_5_apples * 100:.10f}%")
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
probability_apple = 0.6
probability_orange = 0.4

# Number of trials (fruits picked)
n = 20

# Generate x values (number of apples)
x_values = np.arange(0, n+1)

# Calculate the corresponding probabilities using the binomial distribution
probabilities = binom.pmf(x_values, n, probability_apple)

# Plot
plt.bar(x_values, probabilities, color='blue', alpha=0.7, label='Binomial PDF')
plt.title('Binomial Distribution - PDF')
plt.xlabel('Number of Apples')
plt.ylabel('Probability')
plt.xticks(x_values)
plt.legend()
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
# Given average rate of goals
rate_goals = 2.3

# Number of goals we want to find the probability for
k_goals = 5

# Calculate the probability using the Poisson distribution formula
probability_5_goals = (math.exp(-rate_goals) * (rate_goals ** k_goals)) / math.factorial(k_goals)

print(f"Probability of scoring exactly 5 goals: {probability_5_goals:.5f}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
rate_goals = 2.3

range_goals = np.arange(0, 11)

probabilities = poisson.pmf(range_goals, rate_goals)

plt.bar(range_goals, probabilities, color='green', alpha=0.7, label='Poisson PMF')
plt.title('Poisson Probability Distribution for Number of Goals')
plt.xlabel('Number of Goals')
plt.ylabel('Probability')
plt.xticks(x_values)
plt.legend()
plt.show()
````
